# Resolvendo uma EDO


Nosso objetivo será resolver a equação diferencial x'(t) = x.(x - 1) com a condição inicial dada por x(0) = 0.5


```python
import numpy as np
```

```python
from scipy.integrate import odeint
```

```python
import matplotlib.pyplot as plt
```

```python
# Definir a equação que representa a equação diferencial
def model(x, t):
  return x * (x - 1)
```

```python
# Condição inicial
x0 = .5
```

```python
# Pontos de tempo
t = np.linspace(0, 5, 100)
```

```python
# Resolver a equação diferencial
x = odeint(model, x0, t)
```

```python
# Plotando a solução
plt.plot(t, x)
plt.xlabel('Tempo')
plt.ylabel('x(t)')
plt.title('Solução da Equação Diferencial')
plt.show()
```
